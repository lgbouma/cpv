```python
import lightkurve as lk
import numpy as np, matplotlib.pyplot as plt
%matplotlib inline
```

```python
r = lk.search_lightcurve("TIC 289840928")
search_result = lk.search_targetpixelfile("TIC 289840928")
print(r)
print(search_result)
```

```python
lc0 = r[0].download()
tpf0 = search_result[0].download()
lc1 = r[8].download()
tpf1 = search_result[2].download()
```

```python
time = np.array(lc0.time.value)
flux = np.array(lc0.pdcsap_flux.value)
plt.figure(figsize=(15,7))
plt.scatter(time, flux, s=1);
```
